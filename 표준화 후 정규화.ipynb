```python
result_corona = pd.read_csv("checkpoint/result_checkpoint.csv", thousands = ",", index_col = 0)
result_corona.index_names = ["분기"]
result_corona
```

```python
df = pd.DataFrame(data = result_corona[12:])
```

```python
corona = pd.read_csv("./corona2.csv") #코로나 데이터 출력(경수님 것)
```

```python
corona = corona.loc[:,("일일확진자")]
corona = pd.DataFrame(data = corona)
corona.index = ["1Q2020","2Q2020","3Q2020","4Q2020","1Q2021","2Q2021","3Q2021"]
corona
```

```python
df_m = pd.concat([df,corona], ignore_index=True, axis=1)
df_m.columns = ["삼사모바일매출","삼사PC매출","평균기온","마케팅비용","게임시장규모","코로나확진자"]
df_m = df_m.loc[:, ("삼사모바일매출", "삼사PC매출", "마케팅비용", "게임시장규모", "코로나확진자")]
df_m
# 데이터 병합 및 컬럼명 수정
```

```python
plt.figure(figsize=(10, 8))
plt.boxplot(df_m)
plt.xticks(np.arange(5)+1, labels = df_m.columns)
```

## 표준화

```python
# 원래 데이터 표준, 분산 값 확인
print("평균 값")
print(df_m.mean())
print("분산 값")
print(df_m.var())
```

```python
from sklearn.preprocessing import StandardScaler
# StandardScaler 객체 생성
scaler = StandardScaler()
# StandardScaler로 데이터 셋 변환, fit(), transform() 호출
scaler.fit(df_m)
result_scaled = scaler.transform(df_m)

df_standard = pd.DataFrame(data = result_scaled, columns = df_m.columns)

print("평균 값")
print(df_standard.mean())
print("분산 값")
print(df_standard.var())
```

```python
plt.figure(figsize=(10, 8))
plt.boxplot(df_standard)
plt.xticks(np.arange(5)+1, labels = df_standard.columns)
```

## 정규화

```python
from sklearn.preprocessing import MinMaxScaler         

scaler = MinMaxScaler()
scaler.fit(df_standard)
result_scaled = scaler.transform(df_standard)

df_normed = pd.DataFrame(data = result_scaled, columns = df_standard.columns)

print("변수들의 최솟값: \n", df_normed.min())
print("-----------------------------------------------------")
print("변수들의 최댓값: \n", df_normed.max())
```
